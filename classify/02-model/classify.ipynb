```python
# Use all extensions:
# tokenize_folder(path, [])
# This works, but TextDataLoaders.from_folders I can't get
# to work with all extensions.
tokenize_folder(path)
```

```python
dls = TextDataLoaders.from_folder(path, valid_pct=0.15, extensions=[], bs=64)
```

```python
dls.show_batch(max_n=4)
```

```python
# learn = language_model_learner(
#    dls, AWD_LSTM, drop_mult=0.3, 
#    metrics=[accuracy, Perplexity()]).to_fp16()

#learn = language_model_learner(
#    dls, AWD_LSTM, drop_mult=0.3, 
#    metrics=[accuracy, Perplexity()])

#learn = text_classifier_learner(dls, AWD_LSTM, metrics=[accuracy, Perplexity()])



learn = text_classifier_learner(dls, AWD_LSTM, drop_mult=0.5, 
                                metrics=accuracy).to_fp16()

```

```python
learn.fit_one_cycle(1, 2e-2)
```

```python
learn.freeze_to(-2)
learn.fit_one_cycle(1, slice(1e-2/(2.6**4),1e-2))
```
